```python
"""
A nuestra funcín de costo le entran:
par: son los parámetros por ahora solo son dos o0 y o1 en ese orden,
car: nuestro vector numpy.ndarray, es X
et: nuestro vector de etiquetas es Y
"""
def costo(par,X,Y):
  n1=len(X)
  n2=len(Y)
  if(not (n1 == n2)):
    print("ERROR")
  n=n1
  o0=par[0]
  o1=par[1]
  J=(1/(2*n))*np.sum( np.power( ( (o0+o1*X)-(Y) ) ,2) )
  return J
```

```python
lista_par=[]
lista_costos=[]

for o0 in np.arange(-2256.37,-2256.35,0.001):
  #for 2 para iterar en o1
  for o1 in np.arange(7756.30,7756.60,0.001):
    #tomar o0 y o1 y meterlos en un array tupla
    parametros=(o0,o1)
    #prueban su hipotesis usando las funcion de costo que acaban de crear
    lista_costos.append(costo(parametros,X,Y))
    #Guardan el valor del costo y los parámetros con los que se obtuvo ese costo
    lista_par.append(parametros)
    #print(lista_costos)
    #print(lista_par)
# miran cual conjunto de parámetros tuvo el mejor costo. 
# AYUDA miran la funcion ARGMIN del paquete np "numpy"
```

```python
plt.plot(lista_costos)
plt.show()
```

```python
print(np.min(lista_costos)) 
```

```python
par=lista_par[np.argmin(lista_costos)]
print(par)
```
